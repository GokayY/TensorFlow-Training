```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import tensorflow as tf
```

```python
W_true = 2
b_true = 0.5
```

```python
x = np.linspace(0, 3, 130)
y = W_true * x + b_true + np.random.randn(*x.shape) * 0.5
```

```python
plt.figure(figsize = (10,8))
plt.scatter(x,y)

plt.xlabel('x')
plt.ylabel('y')
plt.title('Training Data')
plt.show()
```

```python
class LinearModel:
    
    def __init__(self):
        self.weight = tf.Variable(np.random.randn(), name = 'W')
        self.bias = tf.Variable(np.random.randn(), name = 'b')
    
    def __call__(self,x):
        return self.weight * x + self.bias
```

```python
def loss(y,y_pred):
    return tf.reduce_mean(tf.square(y-y_pred))
```

```python
def train(linear_model, x, y, lr=0.01):
    with tf.GradientTape() as tape:
        y_pred = linear_model(x)
        current_loss = loss(y,y_pred)
    
    d_weight, d_bias = tape.gradient(current_loss, 
                                    [linear_model.weight, linear_model.bias])
    
    linear_model.weight.assign_sub(lr * d_weight)
    linear_model.bias.assign_sub(lr * d_bias)
```

```python
model1 = LinearModel()

weights, biases = [], []
epochs = 100

lr = .15
```

```python
for epoch_count in range(epochs):
    weights.append(model1.weight.numpy())
    biases.append(model1.bias.numpy())
    
    real_loss = loss(y, model1(x))
    train(model1,x,y, lr=lr)
    
    print(f'Epoch count {epoch_count}: Loss value: {real_loss.numpy()}')
```

```python
plt.figure(figsize = (10,8))
plt.plot(range(epochs), weights, 'r', range(epochs), biases, 'b')

plt.plot([W_true] * epochs, 'r--', [b_true] * epochs, 'b--')

plt.legend(['W', 'b', 'true W', 'true b'])
plt.show()
```

```python
model1.weight.numpy(), model1.bias.numpy()
```

```python
rmse = loss(y, model1(x))

rmse.numpy()
```

```python
plt.figure(figsize = (10,8))

plt.plot(x,y, 'ro', label =  'Original Data')
plt.plot(x,model1(x), label = 'Fitted Line')

plt.title('Linear Regression')
plt.legend()
plt.show()
```

#### SimpleLinearRegression with Sequential Model

```python
from tensorflow import keras
from tensorflow.keras import layers
```

```python
x.shape, y.shape
```

```python
x = pd.DataFrame(x, columns=['x'])
y = pd.DataFrame(y, columns=['y'])
```

```python
x.head()
```

```python
y.head()
```

```python
model2 = keras.Sequential((layers.Dense(1, input_shape = (1,), activation = 'linear')))

optimizer = tf.keras.optimizers.SGD(learning_rate = 0.001)

model2.compile(loss = 'mse', metrics = ['mse'], optimizer = optimizer)
```

```python
model2.fit(x,y, epochs = 100)
```

```python
y_pred = model2.predict(x)
```

```python
plt.figure(figsize = (10,8))
plt.scatter(x,y, c='blue', label = 'Original Data')
plt.plot(x,y_pred, color = 'r', label= 'Fitted Line')

plt.title('Linear Regression')
plt.legend()
plt.show()
```
